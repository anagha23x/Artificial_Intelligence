```python
import gym
import time
import matplotlib.pyplot as plt
import random
```

```python
env = gym.make("CartPole-v1",render_mode = "human")
```

```python
observations = list()
rewards = list()
rew_ep = dict()
for j in range(10):
    env.reset()
    random_sample = env.action_space.sample()
    obs,reward,done, done2,info = env.step(random_sample)
    for i in range(100):
        if obs[2]<0:
            obs,reward,done, done2,info = env.step(0)
        else:
            obs,reward,done, done2,info = env.step(1)
        observations.append(obs)
        rewards.append(reward)
        env.render()
        done = done + done2
        if done:
            print(done)
            print(done2)
            env.reset()
            break
    rew_ep[j] =sum(rewards)
    rewards.clear()
env.close()
```

```python
plt.bar(rew_ep.keys(),rew_ep.values())
```
